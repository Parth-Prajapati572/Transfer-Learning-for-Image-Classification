```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import (ResNet50, ResNet101, EfficientNetB0, VGG16, resnet50, resnet, efficientnet, vgg16)
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.regularizers import l2
from sklearn.metrics import classification_report, roc_auc_score, precision_recall_fscore_support
import zipfile
```

```python
train_dir = '../data/seg_train/seg_train'
test_dir = '../data/seg_test/seg_test'
```

```python
IMAGE_SIZE = (150, 150)
BATCH_SIZE = 10
EPOCHS = 60

preprocessing_functions = {
    "ResNet50": resnet50.preprocess_input,
    "ResNet101": resnet.preprocess_input,
    "EfficientNetB0": efficientnet.preprocess_input,
    "VGG16": vgg16.preprocess_input
}

models = {
    "ResNet50": ResNet50(weights='imagenet', include_top=False, input_shape=(150, 150, 3)),
    "ResNet101": ResNet101(weights='imagenet', include_top=False, input_shape=(150, 150, 3)),
    "EfficientNetB0": EfficientNetB0(weights='imagenet', include_top=False, input_shape=(150, 150, 3)),
    "VGG16": VGG16(weights='imagenet', include_top=False, input_shape=(150, 150, 3))
}
```

```python
history_dict = {}
evaluation_metrics = {}

for model_name, base_model in models.items():

    preprocess_input = preprocessing_functions[model_name]
    def custom_preprocessing_train(image):
        image = preprocess_input(image)
        image = tf.image.random_contrast(image, lower=0.8, upper=1.2)
        return image

    def custom_preprocessing_test(image):
        return preprocess_input(image)

    train_datagen = ImageDataGenerator(
        preprocessing_function=custom_preprocessing_train,
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,        
        validation_split=0.2
        )
    test_datagen = ImageDataGenerator(preprocessing_function=custom_preprocessing_test)
    train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=IMAGE_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='training'
    )
    validation_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=IMAGE_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='validation'
    )
    test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=IMAGE_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=False
    )
    print(f"\nTraining model: {model_name}")
    base_model.trainable = False
    inputs = tf.keras.Input(shape=(150, 150, 3))
    x = base_model(inputs, training=False)
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = Dense(64, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = Dropout(0.2)(x)
    x = Dense(32, activation='relu', kernel_regularizer=l2(0.001))(x)
    outputs = Dense(len(train_generator.class_indices), activation='softmax', kernel_regularizer=l2(0.001))(x)
    model = Model(inputs, outputs)

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=5e-4),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    early_stopping = EarlyStopping( monitor='val_loss', patience=50, restore_best_weights=True)

    history = model.fit(
        train_generator,
        validation_data=validation_generator,
        epochs=EPOCHS,
        callbacks=[early_stopping],
        verbose=1
    )
    history_dict[model_name] = history

    train_loss, train_accuracy = model.evaluate(train_generator, verbose=0)
    y_train_pred = model.predict(train_generator)
    y_train_true = train_generator.classes
    y_train_pred_classes = np.argmax(y_train_pred, axis=1)
    train_precision, train_recall, train_f1, _ = precision_recall_fscore_support(
        y_train_true, y_train_pred_classes, average='weighted'
    )
    train_auc = roc_auc_score(
        tf.keras.utils.to_categorical(y_train_true, len(train_generator.class_indices)),
        y_train_pred, multi_class='ovr'
    )
    val_loss, val_accuracy = model.evaluate(validation_generator, verbose=0)
    y_val_pred = model.predict(validation_generator)
    y_val_true = validation_generator.classes
    y_val_pred_classes = np.argmax(y_val_pred, axis=1)
    val_precision, val_recall, val_f1, _ = precision_recall_fscore_support(
        y_val_true, y_val_pred_classes, average='weighted'
    )
    val_auc = roc_auc_score(
        tf.keras.utils.to_categorical(y_val_true, len(validation_generator.class_indices)),
        y_val_pred, multi_class='ovr'
    )
    test_loss, test_accuracy = model.evaluate(test_generator, verbose=0)
    y_test_pred = model.predict(test_generator)
    y_test_true = test_generator.classes
    y_test_pred_classes = np.argmax(y_test_pred, axis=1)
    test_precision, test_recall, test_f1, _ = precision_recall_fscore_support(
        y_test_true, y_test_pred_classes, average='weighted'
    )
    test_auc = roc_auc_score(
        tf.keras.utils.to_categorical(y_test_true, len(test_generator.class_indices)),
        y_test_pred, multi_class='ovr'
    )

    evaluation_metrics[model_name] = {
        "train": {
            "precision": train_precision,
            "recall": train_recall,
            "f1_score": train_f1,
            "auc": train_auc,
            "accuracy": train_accuracy
        },
        "validation": {
            "precision": val_precision,
            "recall": val_recall,
            "f1_score": val_f1,
            "auc": val_auc,
            "accuracy": val_accuracy
        },
        "test": {
            "precision": test_precision,
            "recall": test_recall,
            "f1_score": test_f1,
            "auc": test_auc,
            "accuracy": test_accuracy
        }
    }

    plt.figure()
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{model_name} Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

    plt.figure()
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{model_name} Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```

```python
print("\nFinal Evaluation Metrics:")
for model_name, metrics in evaluation_metrics.items():
    print("*********************")
    print(f"Model: {model_name}")
    print("*********************")

    for dataset, dataset_metrics in metrics.items():
        print(f"  \n{dataset.capitalize()} Metrics:\n")
        for metric_name, value in dataset_metrics.items():
            print(f"    {metric_name.capitalize()}: {value:.4f}\n")
```
